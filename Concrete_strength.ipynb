```python
# From the above graphs , we can observe that  there is no linear relationship between the dependent and the independent features . 
# So a Linear regression model might not be suitable
```

```python
df_linear=df.copy()
```

```python
df_linear.head()
```

```python
# Checking for outliers
```

```python
for i in df_linear.columns:
    if df_linear[i].dtype!='object':
        sns.boxplot(df_linear[i])
        plt.show()
```

```python
# Outlier Treatment
def remove_outlier(cols):
    sorted(cols)
    Q1,Q3=cols.quantile([0.25,0.75])
    IQR=Q3-Q1
    LW=Q1-(1.5*IQR)
    UW=Q3+(1.5*IQR)
    return LW,UW
```

```python
columns=['Cement','Slag','Fly ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate','Age(days)']
for i in columns:
    LW,UW=remove_outlier(df_linear[i])
    df_linear[i]=np.where(df_linear[i]<LW,LW,df_linear[i])
    df_linear[i]=np.where(df_linear[i]>UW,UW,df_linear[i])
```

```python
for i in df_linear.columns:
    if df_linear[i].dtype!='object':
        sns.boxplot(df_linear[i])
        plt.show()
```

```python
# Data Preprocessing
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
```

```python
for i in columns:
    df_linear[[i]]=SC.fit_transform( df_linear[[i]])
```

```python
df_linear.head()
```
